```python
# Calculation of rental property ROI (return on investment)

class Income:
   
    # Class attribute
    purchase_price = float(input("Enter the rental property purchase price: $"))
    total_monthly_income = 0

    rental_income = int(input("Enter monthly rental income: $"))
    laundry_income = int(input("Enter laundry fees for this property\
 (if applicable): $"))
    storage_income = storage = int(input("Enter storage fees for this property\
 (if applicable):  $"))
    misc_income = int(input("Enter additional miscellaneous fees\
 (if applicable): $"))

    total_monthly_income = rental_income + laundry_income + storage_income \
          + misc_income
    print("___________________________________________")
    print( f"Total Monthly Income For This Property = ${total_monthly_income}")

    print("___________________________________________")
    print("How much does it cost to own your property?")
    print("Enter monthly expenses below (if applicable). ")


class Expenses:
    total_monthly_expenses = 0

    insurance = int(input("Insurance fee(s): $"))
    hoa = int(input("HOA fee: $"))
    utilities = int(input("Total amount landlord pays for utility fees\
 (e.g. electric/water/sewer/gas): $"))
    maintenance = int(input("Lawn/Snow fee(s): $"))
    small_repairs = int(input("How much money do you save each month\
 for small repairs? $"))
    cap_ex = int(input("How much money do you save each month for big\
 repairs? (e.g. new roof, water heater) $"))
    property_mgmt = int(input("Cost for property management: $"))
    
    print("\nCalculating Monthly Tax Fee: ")
    print("___________________________________________")
        # user inputs tax rate, converted to a float
    tax_rate = float(input("Enter tax rate (e.g. 8.2):  %"))
    tax_calc = float(tax_rate / 10) * float(Income.purchase_price)\
              /100 
    tax_fee = int((tax_calc) / 12)
    print((f'Tax fee = ${tax_fee}'))
        
    print("\nCalculating Monthly Mortgage Fee: ")
    print("___________________________________________")
    loan_amount = int(input("Enter principal loan amount: $"))
    mortgage_interest = float(input("Enter annual interest rate:  %"))
    loan_term = int(input("Enter the number of years in your \
loan term: "))
        
        
        #vacancy calculations
    vacancy = 0.05 * Income.rental_income
        # print(self.vacancy)
      
        # #mortgage calculations
        # calculates monthly interest rate
    monthly_mortgage_interest = float(mortgage_interest/100) / 12
        # r(1 + r) to the nth power
    top_divide = float(monthly_mortgage_interest * 
                    (1 + monthly_mortgage_interest) ** (loan_term * 12))
        #(1 + r) to the nth power - 1
    bottom_divide = float((1 + monthly_mortgage_interest) ** \
                                   (loan_term * 12) - 1)
    monthly_mortgage_fee = int(loan_amount * top_divide /\
                                           bottom_divide)
    print(f'Monthly Mortagage Fee: ${monthly_mortgage_fee}')
       
        # calculating monthly expenses total
    total_monthly_expenses = int(tax_fee + insurance + \
            utilities + hoa + maintenance + small_repairs \
                + cap_ex + property_mgmt + vacancy\
                    + monthly_mortgage_fee) 
        # print(self.total_monthly_expenses)        

           
class CashFlow():

    total_monthly_cash_flow = int(Income.total_monthly_income - \
            Expenses.total_monthly_expenses)
    # print(total_monthly_cash_flow)
    

class CashOnCashROI():

   print("\nCalculating Cash on Cash Return on Investment: ")
   print("___________________________________________")
   dwn_pymt = int(input("Enter down payment: $"))
   closing_costs = int(input("Enter closing cost: $"))
   rehab_budget = int(input("Rehab reapir amount: $"))
   miscellaneous = int(input("Miscellaneous fees: $"))
   total_investment = dwn_pymt + closing_costs + rehab_budget + miscellaneous
   print(f'Total Investment: ${total_investment}')

   annual_cash_flow = int(CashFlow.total_monthly_cash_flow * 12)
   cash_on_cash_roi = float(annual_cash_flow /total_investment) * 100
   print(f'Cash on Cash ROI = %{cash_on_cash_roi:.2f}')
    

# Income()
# Expenses()
# CashFlow()
#CashonCashROI()
```
